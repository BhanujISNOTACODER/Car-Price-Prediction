## Model

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.pipeline import make_pipeline
from sklearn.compose import make_column_transformer
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
cars.Make = cars.Make.apply(lambda X:X.replace("-",""))
```

```python
inp = cars[['Year','Mileage','Make','Model']]
op = cars[['Price']]
```

```python
inp
```

```python
op
```

```python
ohe = OneHotEncoder()
ohe.fit(inp[["Year","Make","Model"]])
ohe.categories_
```

```python
col_transform = make_column_transformer((OneHotEncoder(categories=ohe.categories_),["Year","Make",'Model']),remainder="passthrough")
```

```python
lr = LinearRegression()
pipe = make_pipeline(col_transform,lr)
```

```python
inp_train,inp_test,op_train,op_test = train_test_split(inp,op,test_size=0.2)
```

```python
pipe.fit(inp_train,op_train)
```

```python
pred = pipe.predict(inp_test)
```

```python
from sklearn.metrics import r2_score
```

```python
scores=[]
for i in range(1000):   
    inp_train,inp_test,op_train,op_test = train_test_split(inp,op,test_size=0.2,random_state=541)
    pipe.fit(inp_train,op_train)
    pred = pipe.predict(inp_test)
    r2 = r2_score(op_test,pred)
    scores.append(r2)
scores
```

```python
np.argmax(scores)
```
